```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'], 'key': ['K0', 'K1', 'K2']})
df2 = pd.DataFrame({'B': ['B0', 'B1', 'B2'], 'key': ['K0', 'K1', 'K3']})
merged = pd.merge(df1, df2, on='key', how='inner')
print(merged)
```

```python
date_rng = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
time_series = pd.DataFrame(date_rng, columns=['date'])
time_series['day_of_week'] = time_series['date'].dt.day_name()
print(time_series)
```

```python
data = {'date': ['2023-01-01', '2023-01-01', '2023-01-02'],
        'product': ['A', 'B', 'A'],
        'sales': [100, 150, 200]}
df = pd.DataFrame(data)
pivot_table = df.pivot_table(index='date', columns='product', values='sales', aggfunc='sum')
print(pivot_table)
```

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'San Francisco', 'Los Angeles']
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

selected_columns = df[['Name', 'Age']]
print("\nSelected Columns:")
print(selected_columns)

filtered_df = df[df['Age'] > 30]
print("\nFiltered DataFrame:")
print(filtered_df)

sorted_df = df.sort_values(by='Age', ascending=False)
print("\nSorted DataFrame:")
print(sorted_df)
```

```python
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C']
values1 = [10, 15, 8]
values2 = [5, 10, 12]

width = 0.35
x = range(len(categories))

plt.bar(x, values1, width, label='Value Set 1')
plt.bar(x, values2, width, label='Value Set 2', bottom=values1)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 8]

plt.fill_between(x, y, alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Area Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 8]
z = [2, 3, 1, 5, 7]

ax.scatter(x, y, z)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('3D Scatter Plot')
plt.show()
```
